### DATA CLEANING AND PREPARATION FOR SAMPLING

```python
#Checking for null Value in Dataset,if True remove Null or empty rows

df.isnull().values.any()
```

```python
#removing Null rows

df['Type 2'] = df['Type 2'].replace('-', np.nan)
df = df.dropna(axis=0, subset=['Type 2'])
```

```python
#dimension of data
df.shape
```

### RANDOM SAMPLING

```python
sample1=df.sample(n=50)
print(sample1)
```

```python
# dimension of Data-sample
sample1.shape
```

### SYSTEMATIC SAMPLING

```python
def systematic_sampling(df, step):
 
    indexes = np.arange(0, len(df), step=step)
    systematic_sample = df.iloc[indexes]
    return systematic_sample

# Obtain a systematic sample and save it in a new variable
systematic_sample = systematic_sampling(df,8)
a=systematic_sample.head(50)
# View sampled data frame
display(a)
```

```python
#dimension of data-sample
a.shape
```

### STRATIFIED SAMPLING

```python
n=51
df.groupby('Type 2', group_keys=False).apply(lambda x: x.sample(int(np.rint(n*len(x)/len(df))))).sample(frac=1).reset_index(drop=True)
```
