#Uni-Variate Relationship

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import seaborn as sn
```

#VISUALIZING UNIVARIATE CONTINUOUS DATA 

UNIVARIATE SCATTER PLOT

```python
df=pd.read_csv('/content/ChiSqare.csv')
plt.scatter(df.index,df['Height'])
```

```python
sn.scatterplot(x=df.index,y=df['Height'],hue=df['Shoe Size'])
```

```python
plt.figure(figsize=(6,6))
plt.title('length')
plt.xlabel('Height', fontsize=20)
plt.ylabel('Shoe Size', fontsize=20)
plt.plot(df.index,df['Height'],markevery=1,marker="d")
plt.show()
```

```python
sn.set(rc={'figure.figsize':(7,7)})
sn.set(font_scale=1.5)
```

```python
fig=sn.lineplot (x=df.index, y=df['Height'],markevery=1,marker="d",data=df, hue=df['Superhero'])
fig.set(xlabel="index")
```

```python
sn.stripplot (x=df['Bfast Calories'])
sn.set(rc={'figure.figsize':(2,2)})
```

```python
sn.stripplot(x=df['Happy'],y=df['Phone Time'])
sn.set(rc={'figure.figsize':(8,8)})
```

```python
sn.set (rc={'figure.figsize':(5,5)})
sn.swarmplot (x=df['Funny'])
```

```python
sn.set (rc={'figure.figsize':(10,5)})
sn.swarmplot (x=df['College'])
```

```python
plt.figure(figsize=(5,5))
df['Social Media'].plot(kind="density")
```

```python
sn.set(rc={'figure.figsize':(5,5)})
sn.kdeplot (df['Happy'] ,shade=True)
```

#**Bivariate Relationship**

```python

sn.countplot(x='Funny',data=df)
```

```python
sn.barplot(x='Happy',y='Funny',data=df)
sn.set (rc={'figure.figsize':(10,8)})
```

```python
sn.boxplot(x='Stat Pre',y='Stat Post',data=df,palette='rainbow')
```

```python
sn.violinplot(x='Stat Pre',y='Stat Post',data=df,palette='rainbow')
```

```python
sn.stripplot(x='Stat Pre',y='Stat Post',data=df,palette='rainbow')
sn.swarmplot(x='Stat Pre',y='Stat Post',data=df,palette='rainbow')
```

```python
g = sn.pairplot(df,diag_kind="kde", markers="+",
                 plot_kws=dict(s=50, edgecolor="b", linewidth=1),
                 diag_kws=dict(shade=True))
```

```python
g = sn.PairGrid(df)
g = g.map_upper(sn.scatterplot)
g = g.map_lower(sn.kdeplot, colors="C0")
g = g.map_diag(sn.kdeplot, lw=2)
```
